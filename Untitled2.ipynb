```python
# Titanic Survivability Analysis

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set(style="whitegrid")
%matplotlib inline

# Load the Dataset
df = pd.read_csv("Project_1/data/titanic.csv")
df.head()
```

```python
for col in ["Survived", "Pclass", "SibSp", "Parch"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce").astype("Int64")
```

```python
def show_df(df, title):
    try:
        display(HTML(f"<h3>{title}</h3>" + df.to_html(index=False)))
    except Exception:
        print(f"\n## {title}\n")
        print(df)
```

```python
show_df(pd.DataFrame({"Rows":[len(df)], "Columns":[len(df.columns)]}), "Dataset Shape")
show_df(pd.DataFrame(df.dtypes, columns=["dtype"]).reset_index().rename(columns={"index":"column"}), "Column dtypes")
```

```python
missing_info = df.isna().sum().reset_index()
missing_info.columns = ["column", "missing_count"]
missing_info["missing_pct"] = (missing_info["missing_count"] / len(df) * 100).round(2) if len(df) else 0
show_df(missing_info, "Missing Values")
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if num_cols:
    show_df(df[num_cols].describe().T, "Numeric Summary")
```

```python
def safe_value_counts(series):
    vc = series.value_counts(dropna=False).sort_index()
    vc.index = vc.index.map(lambda x: "NaN" if (isinstance(x, float) and math.isnan(x)) else str(x))
    return vc
```

```python
def plot_bar(series, title, xlabel, ylabel="Count", rotate=0):
    vc = safe_value_counts(series)
    plt.figure()
    plt.bar(vc.index, vc.values)
    plt.title(title); plt.xlabel(xlabel); plt.ylabel(ylabel)
    if rotate: plt.xticks(rotation=rotate, ha="right")
    plt.tight_layout(); plt.show()
```

```python
def plot_hist(series, title, xlabel, bins=30, logx=False):
    clean = series.dropna()
    plt.figure()
    if logx:
        clean = clean[clean > 0]
        plt.hist(np.log10(clean), bins=bins)
        plt.xlabel(f"log10({xlabel})")
    else:
        plt.hist(clean, bins=bins)
        plt.xlabel(xlabel)
    plt.title(title); plt.ylabel("Frequency")
    plt.tight_layout(); plt.show()
```

```python
import math
for col, (title, rotate) in {
    "Survived": ("Distribution: Survived (0/1)", 0),
    "Pclass": ("Distribution: Passenger Class", 0),
    "Sex": ("Distribution: Sex", 0),
    "Embarked": ("Distribution: Embarked Port", 0),
    "SibSp": ("Distribution: Siblings/Spouses Aboard", 0),
    "Parch": ("Distribution: Parents/Children Aboard", 0),
}.items():
    if col in df.columns:
        plot_bar(df[col], title, col, rotate=0)
```

```python
if "Age" in df.columns:
    plot_hist(df["Age"], "Distribution: Age", "Age", bins=20)
```

```python
plot_hist(df["Fare"], "Distribution: Fare", "Fare", bins=30, logx=False)
plot_hist(df["Fare"], "Distribution: Fare (log10 scale)", "Fare", bins=30, logx=True)
```

```python
df["FamilySize"] = (df["SibSp"].fillna(0) + df["Parch"].fillna(0) + 1) if {"SibSp","Parch"}.issubset(df.columns) else np.nan
df["HasCabin"] = df["Cabin"].notna().astype("Int64") if "Cabin" in df.columns else np.nan
df["Title"] = df["Name"].str.extract(r",\s*([^\.]+)\.", expand=False).str.strip() if "Name" in df.columns else np.nan
```

```python
df["AgeBand"] = pd.cut(df["Age"], [0,12,18,30,50,120], labels=["0-12","13-18","19-30","31-50","51+"], include_lowest=True)
```

```python
q = min(4, max(1, df["Fare"].nunique()))
df["FareBand"] = pd.qcut(df["Fare"], q=q, duplicates="drop")
```

```python
group_tables = {}
def rate_table(by_col):
    g = df.groupby(by_col, dropna=False)["Survived"].mean().reset_index()
    g = g.rename(columns={"Survived": "Rate"})
    g["Survival_Rate_%"] = (g["Rate"] * 100).round(1)
    g = g.drop(columns=["Rate"])
    g = g.dropna(subset=["Survival_Rate_%"])
    # stringify for safe plotting
    g[by_col] = g[by_col].astype(str).replace({"<NA>": "NaN"})
    return g
```

```python
for col in ["Sex","Pclass","Embarked","FamilySize","HasCabin","Title","AgeBand","FareBand"]:
        if col in df.columns and df[col].notna().any():
            tbl = rate_table(col)
            if len(tbl):
                group_tables[col] = tbl
                show_df(tbl, f"Survival Rate by {col}")
                plt.figure()
                plt.bar(tbl[col].astype(str), tbl["Survival_Rate_%"])
                plt.title(f"Survival Rate by {col}")
                plt.xlabel(col)
                plt.ylabel("Survival Rate (%)")
                plt.xticks(rotation=45, ha="right")
                plt.tight_layout()
                plt.show()
```
